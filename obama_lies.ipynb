```python
#first of all import request module in python as it is not in-built. then connect it with desired url using get method and then check it's type using type method.
import requests
r = requests.get('https://danfromsquirrelhill.wordpress.com/2013/08/15/obama-252/')
print(type(r))
```

```python
print(r.text[:500])
```

```python
#create a soup of the url to use inbuilt functions in beautiful soup and use html parser to store it.
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
results = soup.find_all('p')
l_strong = list()
for element in results:
    if element.find('strong') is not None:
        l_strong.append(element)

#only elements from 4-505 are useful in list_strong_elements.
#<p>7) <strong>Waged the biggest war against medical marijuana of any president, which was the opposite of what he had promised<br />
#</strong></p>
```

```python
#making sure whether data is stored or not.
results[20:]
```

```python
#l_strong contains only the strong elements i.e only the lies.
l_strong[4:25]
```

```python
#using get_text method to extract data from the strong tags.
l1_strong=list()      
for i in l_strong[4:506]:
    l1_strong.append(i.get_text(strip=True))
print(l1_strong)
print(len(l1_strong))
```

```python
#agains creating soup to extreact link.
s1=BeautifulSoup(str(results),'lxml')     
l_href=[]
for x in s1.find_all('a'):
    l_href.append(x.get('href'))
l_href1=l_href[5:1133]
#only the links from 5 to 1133 are useful.   
```

```python
#checking the length of links list.
len(l_href1)
```

```python
l_href1[5:1133]
```

```python
#storing indexes in index and lies in lies and links in l1_String...also as you can see data at 477 and 110 is disorganized so we need to remove it using pop method in list. 
indexes=list()
lies=list()
l1_strong.pop(477)
l1_strong.pop(110)
for i in l1_strong:
    l1=i.split(")")
    indexes.append(l1[0])
    lies.append(l1[1])
    

print(indexes)

    
```

```python
#checking lies and it's length.
print(lies)
print(len(lies))
```

```python
print(len(lies))
```

```python
print(len(indexes))
```

```python
#zipping the indexes and list using list zip method.
ll=list(zip(indexes,lies,l_href))
print(ll)
```

```python
#writing the output data in html file and storing it in same dreictory
with open('final.html','w') as f:
    f.write("<html><head><style>td:hover{background-color:#00FFFF;}th{background-color:red;} th,td{padding: 15px;} table,th,td{border-collapse: collapse; width: 100%;  border: 1px solid black;} body {background-color: #808080;} h1{color: #800080;}</style><head><body><h1 align='center'>Choose the Lie from Table Below For evidence.</h1><table ><tr><th>Index</th><th>Lies</th><th>Evidences</th></tr>")
    for x in ll:
        f.write("<tr><td>%s</td><td>%s</td><td><a href=%s>%s</a></td></tr>" %(x[0],x[1],x[2],x[2]))
    f.write("</table></body></html>")
         
```

```python
#importing pandas library and then making dataframes with lies,indexes and l_href as columns.
import pandas as p    
df=p.DataFrame(ll,columns=['indexes','lies','l_href']) 
print(type(df))
```
